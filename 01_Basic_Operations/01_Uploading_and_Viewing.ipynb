```python
# Import OpenCV library for image processing
import cv2

# Import Matplotlib library for displaying images
import matplotlib.pyplot as plt

# Read the image from the current folder
# Replace "image.jpg" with your image file name
img = cv2.imread("image.jpg")

# OpenCV reads images in BGR (Blue-Green-Red) format
# Convert the image from BGR to RGB format for correct color display
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Display the image using Matplotlib
plt.imshow(img)

# Show the image window
plt.show()
```
