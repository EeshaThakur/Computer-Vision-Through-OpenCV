```python
# RGB to GRAY Conversion
import cv2
import matplotlib.pyplot as plt

# Read image
# OpenCV reads image in BGR format

img = cv2.imread("image.jpg")

# Convert BGR to RGB

# OpenCV uses BGR
# Matplotlib uses RGB

rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
# Convert RGB to GRAY

# Gray Image:
# Pixel = Brightness only

# Example:
# 125

# Shape Changes:

# Before:
# (height, width, 3)

# After:
# (height, width)

# The 3 color channels are combined
# into a single intensity value.

# Gray images require less memory
# and are faster to process.

gray = cv2.cvtColor(rgb, cv2.COLOR_RGB2GRAY)

# Check Shape
print("RGB Shape :", rgb.shape)

print("GRAY Shape:", gray.shape)

# Display Grayscale Image

# cmap='gray'
# tells matplotlib to display
# shades of gray.

plt.imshow(gray, cmap='gray')

plt.axis("off")

plt.title("RGB to Gray")

plt.show()
```

```python
# Program to convert Grayscale image to RGB

import cv2
import matplotlib.pyplot as plt

# Syntax:
# cv2.imread(filename, flag)

# Read image in grayscale mode
# 0 means grayscale image

gray = cv2.imread("image.jpg", 0)

# Syntax:
# cv2.cvtColor(src, code)

# Why Gray to RGB?
#
# Grayscale image has only 1 channel.
# RGB image has 3 channels:
# Red, Green and Blue.
#
# Some OpenCV functions require a 3-channel image.
# Therefore we convert Gray to RGB.

rgb = cv2.cvtColor(gray, cv2.COLOR_GRAY2RGB)

plt.imshow(rgb)
plt.title("GRAY to RGB")
plt.show()
```
