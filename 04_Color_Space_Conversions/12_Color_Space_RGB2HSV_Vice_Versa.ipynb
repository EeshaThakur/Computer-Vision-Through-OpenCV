```python
# Program to convert RGB image to HSV

import cv2
import matplotlib.pyplot as plt

# Read image
img = cv2.imread("image.jpg")

# OpenCV reads image in BGR format
# Convert BGR to RGB for proper color representation

rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Why convert RGB to HSV?
# HSV separates color and brightness
# H = Hue (Color)
# S = Saturation (Color strength)
# V = Value (Brightness)
# Useful for color detection and segmentation

hsv = cv2.cvtColor(rgb, cv2.COLOR_RGB2HSV)

plt.imshow(hsv)
plt.title("RGB to HSV")
plt.show()
```

```python
# Program to convert HSV image to RGB

import cv2
import matplotlib.pyplot as plt

# Read image
img = cv2.imread("image.jpg")

# Convert BGR image to HSV
hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

# Syntax:
# cv2.cvtColor(src, code)

# Convert HSV to RGB
rgb = cv2.cvtColor(hsv, cv2.COLOR_HSV2RGB)

# Display image
plt.imshow(rgb)
plt.title("HSV to RGB")
plt.show()
```
