```python
# Program to convert BGR image to RGB image

import cv2
import matplotlib.pyplot as plt

# Syntax:
# cv2.imread(filename)
# Reads the image

img = cv2.imread("image.jpg")

# Syntax:
# cv2.cvtColor(src, code)
# Converts one color format to another

# BGR -> RGB
# OpenCV reads images in BGR format
# Matplotlib displays images in RGB format

rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(rgb)
plt.title("BGR to RGB")
plt.show()
```

```python
# Program to convert RGB image to BGR image

import cv2
import matplotlib.pyplot as plt

img = cv2.imread("image.jpg")

# Convert BGR to RGB first
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# RGB -> BGR
# Converts RGB image back to OpenCV format

bgr = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)

plt.imshow(cv2.cvtColor(bgr, cv2.COLOR_BGR2RGB))
plt.title("RGB to BGR")
plt.show()
```
