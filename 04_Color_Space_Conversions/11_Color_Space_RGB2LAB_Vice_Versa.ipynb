```python
# Program to convert RGB image to LAB

import cv2
import matplotlib.pyplot as plt

# Syntax:
# cv2.imread(filename)
# Reads the image from the file

img = cv2.imread("image.jpg")

# OpenCV reads images in BGR format by default
# Matplotlib displays images correctly in RGB format
# Therefore, convert BGR to RGB first

# Syntax:
# cv2.cvtColor(src, code)

rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Why convert RGB to LAB?
# LAB separates brightness from color information
# L = Lightness
# A = Green ↔ Red
# B = Blue ↔ Yellow
# LAB is useful for image enhancement and color analysis

lab = cv2.cvtColor(rgb, cv2.COLOR_RGB2LAB)

# Display LAB image
plt.imshow(lab)
plt.title("RGB to LAB")
plt.show()
```
