```python
# Program to convert RGB image to YUV

import cv2
import matplotlib.pyplot as plt

img = cv2.imread("image.jpg")

# OpenCV reads image in BGR format
# Convert BGR to RGB first

rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Why RGB to YUV?
#
# RGB stores colors together.
#
# YUV separates:
#
# Y = Brightness
# U = Blue color information
# V = Red color information
#
# Used in video compression and television broadcasting.

yuv = cv2.cvtColor(rgb, cv2.COLOR_RGB2YUV)

plt.imshow(yuv)
plt.title("RGB to YUV")
plt.axis("off")
plt.show()
```

```python
# Program to convert YUV image to RGB

import cv2
import matplotlib.pyplot as plt

img = cv2.imread("image.jpg")

rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

yuv = cv2.cvtColor(rgb, cv2.COLOR_RGB2YUV)

# Why YUV to RGB?
#
# YUV is useful for processing and compression.
# RGB is needed for normal display.

rgb2 = cv2.cvtColor(yuv, cv2.COLOR_YUV2RGB)

plt.imshow(rgb2)
plt.title("YUV to RGB")
plt.show()
```
