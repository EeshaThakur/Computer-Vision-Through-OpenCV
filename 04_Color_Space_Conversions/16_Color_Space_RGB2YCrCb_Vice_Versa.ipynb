```python
# Program to convert RGB image to YCrCb

import cv2
import matplotlib.pyplot as plt

img = cv2.imread("image.jpg")

rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Why RGB to YCrCb?
#
# Y = Brightness
# Cr = Red difference
# Cb = Blue difference
#
# Used in:
# Face detection
# Video compression
# Image processing

ycrcb = cv2.cvtColor(rgb, cv2.COLOR_RGB2YCrCb)

plt.imshow(ycrcb)
plt.title("RGB to YCrCb")
plt.show()
```

```python
# Program to convert YCrCb image to RGB

import cv2
import matplotlib.pyplot as plt

img = cv2.imread("image.jpg")

rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

ycrcb = cv2.cvtColor(rgb, cv2.COLOR_RGB2YCrCb)

# Why YCrCb to RGB?
#
# YCrCb is mainly used for processing.
# RGB is needed for normal display.

rgb2 = cv2.cvtColor(ycrcb, cv2.COLOR_YCrCb2RGB)

plt.imshow(rgb2)
plt.title("YCrCb to RGB")
plt.show()
```
