```python
# RGB to XYZ Conversion

import cv2
import matplotlib.pyplot as plt
# Read image
# OpenCV reads image in BGR format

img = cv2.imread("image.jpg")

# Convert BGR to RGB
# OpenCV -> BGR
# XYZ conversion below expects RGB input

rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Convert RGB to XYZ

# X = Red-Green information
# Y = Brightness (Luminance)
# Z = Blue information

# XYZ is a device-independent color space

# Used in:
# - Color science
# - Color calibration
# - Color management systems

# Shape remains:
# (height, width, 3)
xyz = cv2.cvtColor(rgb, cv2.COLOR_RGB2XYZ)

# Display XYZ image

# XYZ is not meant for direct human viewing.
# Colors may look unusual.

plt.imshow(xyz)

plt.axis("off")

plt.title("RGB to XYZ")

plt.show()
```

```python
# XYZ to RGB Conversion
import cv2
import matplotlib.pyplot as plt

# Read image
# OpenCV reads image in BGR format

img = cv2.imread("image.jpg")

# Convert BGR to XYZ

# X = Red-Green information
# Y = Brightness (Luminance)
# Z = Blue information

# Creates an XYZ image
xyz = cv2.cvtColor(img, cv2.COLOR_BGR2XYZ)
# Convert XYZ to RGB

# Why?
# XYZ is mainly used for color science
# and color calculations.

# Humans and display libraries prefer RGB.

# Therefore XYZ is converted back to RGB
# for visualization.
rgb = cv2.cvtColor(xyz, cv2.COLOR_XYZ2RGB)
# Display RGB Image

# plt.imshow() expects RGB format

plt.imshow(rgb)

# Hide x and y axes
plt.axis("off")

# Add title
plt.title("XYZ to RGB")

# Show image
plt.show()
```
