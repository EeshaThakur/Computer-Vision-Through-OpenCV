```python
# Program to draw an ellipse on an image

# Import OpenCV library for image processing
import cv2

# Import Matplotlib library for displaying images
import matplotlib.pyplot as plt

# Syntax:
# cv2.imread(filename)
# Reads an image from the specified file

img = cv2.imread("image.jpg")

# Syntax:
# cv2.cvtColor(src, code)
# Converts image from one color space to another

# OpenCV reads images in BGR format
# Matplotlib displays images in RGB format
# Therefore convert BGR to RGB

img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Syntax:
# cv2.ellipse(image, center_coordinates, axesLength,
#             angle, startAngle, endAngle,
#             color, thickness)

# Draw an ellipse on the image
#
# Parameters:
#
# image  = Input image
#
# (100,100)
# Center point of ellipse
# x = 100, y = 100
#
# (80,40)
# Axes lengths
# 80 = Major axis radius
# 40 = Minor axis radius
#
# 0
# Rotation angle of ellipse
# 0 means no rotation
#
# 0
# Starting angle
#
# 360
# Ending angle
#
# 0 to 360 draws a complete ellipse
#
# (255,0,0)
# Ellipse color (Red in RGB)
#
# 3
# Thickness of ellipse boundary

cv2.ellipse(img, (100,100), (80,40), 0, 0, 360, (255,0,0), 3)

# Syntax:
# plt.imshow(image)
# Displays image

plt.imshow(img)

# Syntax:
# plt.title(title_name)
# Adds title to image

plt.title("Ellipse")

# Syntax:
# plt.show()
# Displays output window

plt.show()
```
