```python
# Program to draw a filled circle on an image

# Import OpenCV library for image processing
import cv2

# Import Matplotlib library for displaying images
import matplotlib.pyplot as plt

# Syntax:
# cv2.imread(filename)
# Reads an image from the specified file

img = cv2.imread("image.jpg")

# Syntax:
# cv2.cvtColor(src, code)
# Converts image from one color space to another

img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Syntax:
# cv2.circle(image, center_coordinates, radius, color, thickness)

# Draw a filled circle on the image
# (110,110) -> Center coordinates of the circle
# 60        -> Radius of the circle
# (255,0,0) -> Circle color (Red in RGB)
# -1        -> Fills the circle with the specified color

cv2.circle(img, (110,110), 60, (255,0,0), -1)

# Syntax:
# plt.imshow(image)
# Displays the image

plt.imshow(img)

# Syntax:
# plt.title(title_name)
# Adds a title to the displayed image

plt.title("Filled Circle")

# Syntax:
# plt.show()
# Shows the output window

plt.show()
```
