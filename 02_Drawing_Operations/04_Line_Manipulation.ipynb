```python
# Program to draw a line on an image

# Import OpenCV library for image processing
import cv2

# Import Matplotlib library for displaying images
import matplotlib.pyplot as plt

# Read the image from the current folder
img = cv2.imread("image.jpg")

# Convert the image from BGR to RGB format
# OpenCV reads images in BGR format, while Matplotlib uses RGB
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Draw a line on the image
# (20,20)   -> Starting point of the line
# (200,200) -> Ending point of the line
# (255,0,0) -> Line color (Red in RGB)
# 3         -> Thickness of the line
cv2.line(img, (20, 20), (200, 200), (255, 0, 0), 3)

# Display the image with the drawn line
plt.imshow(img)

# Show the image window
plt.show()
```
