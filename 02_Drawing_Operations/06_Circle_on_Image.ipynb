```python
# Program to draw a circle on an image

# Import OpenCV library for image processing
import cv2

# Import Matplotlib library for displaying images
import matplotlib.pyplot as plt

# Read the image from the current folder
img = cv2.imread("image.jpg")

# Convert the image from BGR to RGB format
# OpenCV reads images in BGR, while Matplotlib displays in RGB
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Draw a circle on the image
# (110,110) -> Center coordinates of the circle
# 60        -> Radius of the circle
# (255,0,0) -> Circle color (Red in RGB)
# 3         -> Thickness of the circle boundary
cv2.circle(img, (110, 110), 60, (255, 0, 0), 3)

# Display the image with the circle
plt.imshow(img)

# Show the image window
plt.show()
```
