```python
# Program to draw a rectangle on an image

# Import OpenCV library for image processing
import cv2

# Import Matplotlib library for displaying images
import matplotlib.pyplot as plt

# Read the image from the current folder
img = cv2.imread("image.jpg")

# Convert the image from BGR to RGB format
# OpenCV reads images in BGR, while Matplotlib displays in RGB
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Draw a rectangle on the image
# (30,30)   -> Top-left corner of the rectangle
# (180,150) -> Bottom-right corner of the rectangle
# (0,255,0) -> Rectangle color (Green in RGB)
# 3         -> Thickness of the rectangle border
cv2.rectangle(img, (30, 30), (180, 150), (0, 255, 0), 3)

# Display the image with the rectangle
plt.imshow(img)

# Show the image window
plt.show()
```
