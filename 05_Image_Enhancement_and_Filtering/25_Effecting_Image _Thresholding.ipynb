```python
import cv2
import matplotlib.pyplot as plt

# Read the image
image = cv2.imread("source.jpg")

if image is None:
    print("Error: Cannot find 'source.jpg'")
else:
    # Apply thresholding
    _, img1 = cv2.threshold(image, 50, 255, cv2.THRESH_BINARY)
    _, img2 = cv2.threshold(image, 50, 255, cv2.THRESH_BINARY_INV)
    _, img3 = cv2.threshold(image, 50, 255, cv2.THRESH_TRUNC)
    _, img4 = cv2.threshold(image, 50, 255, cv2.THRESH_TOZERO)
    _, img5 = cv2.threshold(image, 50, 255, cv2.THRESH_TOZERO_INV)

    # Create subplots (2 rows × 3 columns)
    fig, axes = plt.subplots(2, 3, figsize=(12, 8))

    # Place each image manually (no loops)
    axes[0, 0].imshow(cv2.cvtColor(img1, cv2.COLOR_BGR2RGB))
    axes[0, 0].set_title("Binary")
    axes[0, 0].axis("off")

    axes[0, 1].imshow(cv2.cvtColor(img2, cv2.COLOR_BGR2RGB))
    axes[0, 1].set_title("Binary Inverted")
    axes[0, 1].axis("off")

    axes[0, 2].imshow(cv2.cvtColor(img3, cv2.COLOR_BGR2RGB))
    axes[0, 2].set_title("Truncated")
    axes[0, 2].axis("off")

    axes[1, 0].imshow(cv2.cvtColor(img4, cv2.COLOR_BGR2RGB))
    axes[1, 0].set_title("To Zero")
    axes[1, 0].axis("off")

    axes[1, 1].imshow(cv2.cvtColor(img5, cv2.COLOR_BGR2RGB))
    axes[1, 1].set_title("To Zero Inverted")
    axes[1, 1].axis("off")

    # Hide the unused subplot (last one in 2x3 grid)
    axes[1, 2].axis("off")

    plt.tight_layout()
    plt.show()

```
