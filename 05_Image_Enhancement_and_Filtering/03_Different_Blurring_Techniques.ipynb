```python
# Program to apply different blurring techniques using OpenCV

# Import OpenCV library for image processing
import cv2

# Import Matplotlib library for displaying images
import matplotlib.pyplot as plt

# Syntax:
# cv2.imread(filename)
# Reads an image from the specified file

# Read original image
image_Original = cv2.imread("image.jpg")

# Read image copies for different blurring methods
image_MedianBlur = cv2.imread("image.jpg")
image_GaussianBlur = cv2.imread("image.jpg")
image_BilateralBlur = cv2.imread("image.jpg")

# Convert all images from BGR to RGB
# Why?
# OpenCV reads images in BGR format
# Matplotlib displays images in RGB format

image_Original = cv2.cvtColor(image_Original, cv2.COLOR_BGR2RGB)
image_MedianBlur = cv2.cvtColor(image_MedianBlur, cv2.COLOR_BGR2RGB)
image_GaussianBlur = cv2.cvtColor(image_GaussianBlur, cv2.COLOR_BGR2RGB)
image_BilateralBlur = cv2.cvtColor(image_BilateralBlur, cv2.COLOR_BGR2RGB)

# -----------------------------------------
# Median Blur
# -----------------------------------------

# Syntax:
# cv2.medianBlur(src, ksize)

# Why Median Blur?
# Removes salt-and-pepper noise
# Replaces each pixel with the median value
# Preserves edges better than average blur

# ksize = 9
# Kernel size must be odd

image_MedianBlur = cv2.medianBlur(image_MedianBlur, 9)

# -----------------------------------------
# Gaussian Blur
# -----------------------------------------

# Syntax:
# cv2.GaussianBlur(src, ksize, sigmaX)

# Why Gaussian Blur?
# Smooths image
# Reduces noise and details
# Commonly used before edge detection

# (9,9) = Kernel size
# 10 = Standard deviation

image_GaussianBlur = cv2.GaussianBlur(
    image_GaussianBlur,
    (9,9),
    10
)

# -----------------------------------------
# Bilateral Filter
# -----------------------------------------

# Syntax:
# cv2.bilateralFilter(src, d, sigmaColor, sigmaSpace)

# Why Bilateral Filter?
# Removes noise
# Preserves image edges
# Produces smoother results than Gaussian blur

# d = 9
# Diameter of pixel neighborhood

# sigmaColor = 100
# Controls color similarity

# sigmaSpace = 75
# Controls spatial distance

image_BilateralBlur = cv2.bilateralFilter(
    image_BilateralBlur,
    9,
    100,
    75
)

# -----------------------------------------
# Display Images
# -----------------------------------------

plt.figure(figsize=(12,8))

# Original Image
plt.subplot(2,2,1)
plt.imshow(image_Original)
plt.title("Original Image")
plt.axis("off")

# Median Blur
plt.subplot(2,2,2)
plt.imshow(image_MedianBlur)
plt.title("Median Blur")
plt.axis("off")

# Gaussian Blur
plt.subplot(2,2,3)
plt.imshow(image_GaussianBlur)
plt.title("Gaussian Blur")
plt.axis("off")

# Bilateral Blur
plt.subplot(2,2,4)
plt.imshow(image_BilateralBlur)
plt.title("Bilateral Filter")
plt.axis("off")

plt.show()
```
