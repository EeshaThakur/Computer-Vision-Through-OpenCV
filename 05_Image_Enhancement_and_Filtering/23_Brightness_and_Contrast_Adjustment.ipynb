```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread("image.jpg")

# Line
cv2.line(img, (50,50), (300,300), (0,0,255), 3)

# Rectangle
cv2.rectangle(img, (100,100), (300,250), (0,255,0), 3)

# Circle
cv2.circle(img, (450,200), 80, (255,0,0), 3)

img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img_rgb)
plt.axis("off")
plt.show()
```

```python
# Brightness and Contrast Adjustment using OpenCV

import cv2
import numpy as np

# Read image
image = cv2.imread("image.jpg")

# Create an empty image with same size and type
new_image = np.zeros(image.shape, image.dtype)

# Contrast and Brightness values
contrast = 3.0   # Alpha (Contrast)
bright = 2       # Beta (Brightness)

# Apply contrast and brightness
for y in range(image.shape[0]):
    for x in range(image.shape[1]):
        for c in range(image.shape[2]):
            new_image[y, x, c] = np.clip(
                contrast * image[y, x, c] + bright,
                0,
                255
            )

# Display images
cv2.imshow("Original Image", image)
cv2.imshow("Brightness and Contrast Adjusted Image", new_image)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread("image.jpg")

rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

rgb = rgb.astype(np.float32)/255.0

Y = 0.299*rgb[:,:,0] + 0.587*rgb[:,:,1] + 0.114*rgb[:,:,2]
I = 0.596*rgb[:,:,0] - 0.275*rgb[:,:,1] - 0.321*rgb[:,:,2]
Q = 0.212*rgb[:,:,0] - 0.523*rgb[:,:,1] + 0.311*rgb[:,:,2]

yiq = np.dstack((Y,I,Q))

plt.imshow(Y, cmap="gray")
plt.axis("off")
plt.title("RGB to YIQ")
plt.show()
```
