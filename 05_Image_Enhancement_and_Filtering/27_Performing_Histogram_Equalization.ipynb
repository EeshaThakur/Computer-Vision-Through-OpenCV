```python
# ===================== IMPORT PACKAGES =====================

# Syntax:
# import module_name
# Purpose: Imports OpenCV library for image processing.
import cv2

# Syntax:
# import module_name as alias_name
# Purpose: Imports matplotlib for displaying images.
import matplotlib.pyplot as plt


# ===================== READ IMAGE =====================

# Syntax:
# variable = cv2.imread("image_path")

# Parameter:
# "image_path" -> Name or location of the image file.

# Return Value:
# Returns the image as a NumPy array.
# Returns None if the image is not found.

src = cv2.imread("source.jpg")


# Check whether the image is loaded successfully.
if src is None:
    print("Error: Cannot find 'source.jpg'")

else:

    # ===================== CONVERT TO GRAYSCALE =====================

    # Syntax:
    # output = cv2.cvtColor(src, code)

    # Parameters:
    # src  -> Input image.
    # code -> Type of color conversion.

    # Here:
    # cv2.COLOR_BGR2GRAY converts the image
    # from BGR format to Grayscale.

    src = cv2.cvtColor(src, cv2.COLOR_BGR2GRAY)


    # ===================== HISTOGRAM EQUALIZATION =====================

    # Syntax:
    # output = cv2.equalizeHist(src)

    # Parameter:
    # src -> Input grayscale image.

    # Return Value:
    # Returns the histogram equalized image.

    # Purpose:
    # Improves the contrast of the image by redistributing
    # the pixel intensity values.

    src_eqlzd = cv2.equalizeHist(src)


    # ===================== DISPLAY ORIGINAL IMAGE =====================

    # Syntax:
    # plt.figure(number)
    # Creates a new figure window.

    plt.figure(0)

    # Syntax:
    # plt.imshow(image, cmap='gray')

    # Parameters:
    # image -> Image to display.
    # cmap  -> Color map ('gray' for grayscale images).

    plt.imshow(src, cmap='gray')

    # Syntax:
    # plt.title("Title")
    plt.title("Original Grayscale Image")

    # Removes x-axis and y-axis.
    plt.axis('off')


    # ===================== DISPLAY EQUALIZED IMAGE =====================

    plt.figure(1)
    plt.imshow(src_eqlzd, cmap='gray')
    plt.title("Histogram Equalized Image")
    plt.axis('off')


    # ===================== DISPLAY BOTH FIGURES =====================

    # Syntax:
    # plt.show()

    # Purpose:
    # Displays all figure windows.

    plt.show()
```
