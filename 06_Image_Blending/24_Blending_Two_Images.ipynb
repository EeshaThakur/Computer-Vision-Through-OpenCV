```python
# Program to blend two images using OpenCV

# Import OpenCV library for image processing
import cv2

# Import Matplotlib library for displaying images
import matplotlib.pyplot as plt

# Syntax:
# cv2.imread(filename)
# Reads an image from the specified file

# Read first image
img1 = cv2.imread("image.jpg")

# Read second image
img2 = cv2.imread("image.jpg")

# OpenCV reads images in BGR format
# Convert both images to RGB for correct display in Matplotlib

# Syntax:
# cv2.cvtColor(src, code)

img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)

# Define blending weights

alpha = 0.30
beta = 0.70

# Meaning:
#
# alpha = Weight of first image
# beta  = Weight of second image
#
# 30% of image 1
# 70% of image 2
#
# alpha + beta is usually equal to 1

# Syntax:
# cv2.addWeighted(src1, alpha, src2, beta, gamma)

# Parameters:
#
# src1  = First image
# alpha = Weight of first image
# src2  = Second image
# beta  = Weight of second image
# gamma = Brightness value added to result
#         0.0 means no extra brightness

# Blend both images

final_image = cv2.addWeighted( img1, alpha, img2, beta, 0.0)

# Syntax:
# plt.imshow(image)
# Displays image

plt.imshow(final_image)

# Syntax:
# plt.title(title_name)
# Adds title to image

plt.title("Blended Image")

# Syntax:
# plt.show()
# Displays output window

plt.show()
```
