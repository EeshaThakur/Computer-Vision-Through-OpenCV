```python
# Program to blur an image using Gaussian Blur

# Import OpenCV library for image processing
import cv2

# Import Matplotlib library for displaying images
import matplotlib.pyplot as plt

# Syntax:
# cv2.imread(filename)
# Reads an image from the specified file

img = cv2.imread("image.jpg")

# Syntax:
# cv2.cvtColor(src, code)
# Converts image from one color space to another

# OpenCV reads images in BGR format
# Matplotlib displays images in RGB format
# Therefore convert BGR to RGB

img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Syntax:
# cv2.GaussianBlur(src, ksize, sigmaX)

# Apply Gaussian Blur
#
# Why Gaussian Blur?
# Reduces noise and image details
# Makes the image appear smooth
# Useful before edge detection and image processing
#
# Parameters:
#
# src    = Input image
#
# ksize  = (11,11)
#          Kernel size
#          Width = 11 pixels
#          Height = 11 pixels
#          Larger kernel = More blur
#
# sigmaX = 0
#          Standard deviation in X direction
#          OpenCV automatically calculates it

blur = cv2.GaussianBlur(img, (11,11), 0)

# Syntax:
# plt.imshow(image)
# Displays image

plt.imshow(blur)

# Syntax:
# plt.title(title_name)
# Adds title to image

plt.title("Blurred Image")

# Syntax:
# plt.show()
# Displays output window

plt.show()
```
