```python
# Program to flip an image horizontally

# Import OpenCV library for image processing
import cv2

# Import Matplotlib library for displaying images
import matplotlib.pyplot as plt

# Syntax:
# cv2.imread(filename)
# Reads an image from the specified file

img = cv2.imread("image.jpg")

# Syntax:
# cv2.cvtColor(src, code)
# Converts image from one color space to another

# OpenCV reads images in BGR format
# Matplotlib displays images in RGB format
# Therefore convert BGR to RGB

img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Syntax:
# cv2.flip(src, flipCode)

# Flip the image horizontally
#
# Why flip an image?
# Creates a mirror image effect
# Useful in image augmentation and editing
#
# Parameters:
#
# src      = Input image
#
# flipCode = 1
#            Flips image horizontally
#            (Left ↔ Right)

flip = cv2.flip(img, 1)

# Syntax:
# plt.imshow(image)
# Displays image

plt.imshow(flip)

# Syntax:
# plt.title(title_name)
# Adds title to image

plt.title("Flipped Image")

# Syntax:
# plt.show()
# Displays output window

plt.show()
```
