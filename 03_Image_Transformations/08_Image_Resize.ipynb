```python
# Program to resize an image

# Import OpenCV library for image processing
import cv2

# Import Matplotlib library for displaying images
import matplotlib.pyplot as plt

# Syntax:
# cv2.imread(filename)
# Reads an image from the specified file

img = cv2.imread("image.jpg")

# Syntax:
# cv2.cvtColor(src, code)
# Converts image from one color space to another

img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Syntax:
# cv2.resize(src, dsize)
# Resizes the image to the specified width and height

# Resize the image to 400 × 400 pixels
# (400,400) -> (width, height)
img = cv2.resize(img, (400, 400))

# Syntax:
# plt.imshow(image)
# Displays the image

plt.imshow(img)

# Syntax:
# plt.show()
# Shows the output window

plt.show()
```
