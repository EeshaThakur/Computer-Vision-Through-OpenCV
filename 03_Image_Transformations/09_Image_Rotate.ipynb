```python
# Program to rotate an image by 45 degrees

# Import OpenCV library for image processing
import cv2

# Import Matplotlib library for displaying images
import matplotlib.pyplot as plt

# Syntax:
# cv2.imread(filename)
# Reads an image from the specified file

img = cv2.imread("image.jpg")

# Syntax:
# cv2.cvtColor(src, code)
# Converts image from one color space to another

img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Get the height and width of the image
# shape[:2] returns (height, width)

(h, w) = img.shape[:2]

# Syntax:
# cv2.getRotationMatrix2D(center, angle, scale)
# Creates a rotation matrix

# (w//2, h//2) -> Center of the image
# 45            -> Rotation angle in degrees
# 1             -> Scaling factor

M = cv2.getRotationMatrix2D((w//2, h//2), 45, 1)

# Syntax:
# cv2.warpAffine(src, M, dsize)
# Applies affine transformation using the rotation matrix

# src   -> Input image
# M     -> Rotation matrix
# dsize -> Output image size (width, height)

rot = cv2.warpAffine(img, M, (w, h))

# Syntax:
# plt.imshow(image)
# Displays the image

plt.imshow(rot)

# Syntax:
# plt.show()
# Shows the output window

plt.show()
```
